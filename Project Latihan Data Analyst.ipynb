# Segmentasi Berdasarkan Indeks Kebahagiaan

```python
bins = [0, 60, 70, 80]
labels = ['Rendah', 'Sedang', 'Tinggi']
df['Kategorisasi'] = pd.cut(df['Indeks_kebahagiaan'], bins=bins, labels=labels)
print(df['Kategorisasi'].value_counts())
```

*Menunjukkan bahwa banyak daerah memiliki Indeks Kebahagiaan yang baik*


*ada 25 Provinsi yang memiliki indeks kebahagiaan diatas 70 - lebih dari rentang nilai*


*ada 8 Provinsi yang memiliki indeks kebahagiaan diantara 60-70 - lebih dari rentang nilai*

Kategori "Tinggi" memiliki jumlah terbanyak, dengan 25 provinsi termasuk dalam kategori ini, menunjukkan bahwa sebagian besar provinsi memiliki tingkat kebahagiaan yang tinggi berdasarkan indeks yang telah dihitung.
Kategori "Sedang" terdiri dari 8 provinsi, menunjukkan adanya kelompok provinsi dengan kebahagiaan yang berada pada level menengah.
Tidak ada provinsi yang masuk dalam kategori "Rendah", yang berarti semua provinsi memiliki Indeks Kebahagiaan yang berada di kisaran 60 ke atas.
